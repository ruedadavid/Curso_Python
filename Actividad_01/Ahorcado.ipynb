```python
def palabra_aleatoria():

    import requests
    from bs4 import BeautifulSoup
    from unicodedata import normalize
    
    r = requests.get('https://www.palabrasaleatorias.com/?fs=1&fs2=0&Submit=Nueva+palabra')
    contenido = r.content
    soup = BeautifulSoup(contenido, 'html.parser')
    caja = soup.find('table')
    valor = caja.get_text()
    palabra = valor.replace('\n','').replace('\r',' ').split()[0].lower()
    palabra = normalize("NFKD", palabra).encode("ascii","ignore").decode("ascii")
    return palabra
```
