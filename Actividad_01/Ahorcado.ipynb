```python
def palabra_aleatoria():
    '''
      Función retorna una palabra para ser utilizada como
     parte del juego.

     Parámetros:
     
      <<sin parámetros de ingreso>>
    
     Variables de Retorno:
    
      palabra = entrega una palabra sin acentos o caracteres 
            especiales para que haga parte del juego

     Dependencias:
     
      unicodedata versión 11.0.0 (normalize)
      requests versión 2.22.0
      bs4 versión 4.7.1
     
     Sitio para extracción de palabras
     
      https://www.palabrasaleatorias.com/?fs=1&fs2=0&Submit=Nueva+palabra
            
     Fecha:
     
      2019-09-02. Creación de función y se agregan comentarios detallados

    '''
    # Se importan librerías al espacio de trabajo
    import requests # libreía para manejo de urls
    from bs4 import BeautifulSoup # Librería para extracción de textos (tags)
    from unicodedata import normalize # Librería conversión de cadenas de texto
    
    # Declaración e inicialización sitio para consulta de datos
    url = 'https://www.palabrasaleatorias.com/?fs=1&fs2=0&Submit=Nueva+palabra'
    # Obtencion de la url de la página web con el generador de palabras
    r = requests.get(url)
    # Obtencion del contenido de la página web
    contenido = r.content
    # Uso de BeautifulSoup para extraer contenido en formato html
    soup = BeautifulSoup(contenido, 'html.parser')
    # Extracción de la primera tabla disponible en el html
    caja = soup.find('table')
    # Extracción de textos con listado de palabras en la tabla
    valor = caja.get_text()
    #  Transformación de textos disponibles en la página web, eliminando 
    # saltos de línea (\n) y tabuladores (\t), conviertiendo todo a minúsculas
    # y extrayendo la primera palabra en la lista
    palabra = valor.replace('\n','').replace('\r',' ').split()[0].lower()
    # Eliminación de acentos mediante el uso de unicode
    palabra = normalize("NFKD", palabra).encode("ascii","ignore").decode("ascii")
    # entrega una palabra para que haga parte del juego
    return palabra
```

```python
def ingresar_letra():    
    '''
     Función solicita al usuario una letra, evalua que se halla
    ingresado un caracter válido, de lo contrario solicita un 
    nuevo ingreso de valor.

     Parámetros:
     
      <<sin parámetros de ingreso>>
    
     Variables de Retorno:
    
      letra = caracter válido (sin acentos, sin caracteres
            especiales

     Dependencias:
     
      unicodedata versión 11.0.0 (normalize)
     
     Fecha:
     
      2019-09-02. Se optimiza el código eliminando la variable
              continuar, se agregan comentarios detallados para la
              función, se agrega conversión de acentos utilizando
              librería unicodedata y se fuerza a ingresar una sola
              letra
      2019-08-15. Creación de función

    '''
    from unicodedata import normalize
    # Se incializa la variable con una caracter numérico
    letra = '1'
    # Inicial ciclo para solicitar un texto válido
    while not(letra.isalpha()):
        # Se solicita el ingreso de una letra que es convertida a minúscula
        letra = str(input('Ingrese una letra: ')).lower()
        # Al caracter ingresado se le eliminan los acentos
        letra = normalize("NFKD", letra).encode("ascii","ignore").decode("ascii")
        # Se avalua si el caracter ingresado es una letra
        if letra.isalpha() == False:
            # Se entrega mensaje al usuario indicado que el caracter ingresado no válido
            print ("Ingrese una letra (no son válidos números o caracteres especiales")
        # Se evalua si la cadana de texto es un sólo caracter
        elif len(letra)>1:
            #  Se asigna una valor numérico para forzar nueva ejecuación
            # del ciclo de validación
            letra = '1'
            # Se entrega mensaje al usuario indicado que el número caracteres ingresado no válido
            print ("Ingrese únicamente una letra (caracteres especiales no son válidos)")            
    # Retorna caracter válido para iniciar el programa
    return letra
```

```python
def cambiar_letra(cadena_original, cadena_ofuscada, letra_buscada):
    '''
     Función desofusca o descubre un texto según la letra buscada
     
     Parámetros:
     
      cadena_original = cadena de texto para ser buscada
      cadena_ofuscada = cadena de texto ofuscada
      letra_buscada = letra ingresada para ser mostrada en caso
                      de estar presente en texto original (cadena_original)
    
     Variables de Retorno:
    
      cadena_retorno = cadena de texto desofuscada, en donde se muestra
                       la cadena de texto descubierta (desofuscada)

     Dependencias:
     
      <<sin dependencias>>
     
    Fecha:
     
      2019-09-02. Creación de función y se agregan comentarios detallados
    '''
    # Inicialización de lista desofuscada (descubierta)
    cadena_retorno = []
    # Ciclo para evaluar las letras de la cadena de texto original
    for i in range(len(cadena_original)):
        #  Condicional para encontrar la posición de la letra
        # buscada en la cadena de texto original
        if cadena_original[i]==letra_buscada:
            #  Se agrega la letra de la cadena de texto original
            # a la cadena_retorno
            cadena_retorno.append(cadena_original[i])
        else:
            #  Se agrega el caracter ofuscado a la cadena_retorno
            cadena_retorno.append(cadena_ofuscada[i])
    # Se retorna una lista con letras desofuscadas al nivel requerido
    return cadena_retorno
```

```python
def main(cad, opciones=len(IMAGENES)):
    '''
     Función principal ejecuta el juego llamando las funciones necesarias para
     mostrar la interfaz básica y realizar las validaciones.
     
     Parámetros:
     
      cad = cadena de texto para ser buscada
      opciones = Número de vidas disponibles
      
     Variables de Retorno:
    
      <<sin variables de retorno>>

     Dependencias:
     
      <<sin dependencias>>
           
     Fecha:
     
      2019-09-02. Creación de función y se agregan comentarios detallados
    '''
    # Declarar e inicializar cadena de texto ofuscada
    cad2 = '_'*len(cad)
    # Inicio de ciclo que contrala las opciones disponibles hacer el juego
    while (opciones>0):
        # Llamado a función que controla el ingreso de las letras
        dato_ingresado = ingresar_letra()
        # incialización de variable para comparación de aciertos
        cad3 = cad2
        #  Llamado a función que intercambia las letras ofuscadas y letras
        # de la palabra original
        cad2 = cambiar_letra(cad, cad2, dato_ingresado)
        # Transforma la lista entregada en una cadena de texto con menor ofuscación
        cad2 = ''.join(cad2)
        # Impresión de la ASCII Art
        print(IMAGENES[opciones-1])
        # Impresión del estado actual de la cadena de texto ofuscada
        print(cad2)
        #  Condicional evalua si la cadena de texto original es igual
        # a la cadena de texto desofuscada
        if cad2 == cad:
            # Indica que el juego ha finalizado y se ha encontrado la palabra buscada
            print('Felicitaciones ganaste. \nEnoontraste la buscada ({})' .format(cad))
            # Instrucción finaliza el ciclo
            break
        #  Parte del condicional evalua si la cadena ofuscada es igual a la variable
        # de comparación
        elif cad3 == cad2:
            #  Se disminuye una opción porque la letra ingresada no ha ayudado a
            # encontrar la palabra buscada
            opciones -=1
    # Condicional revisa si se han agotado las opciones
    if opciones ==0:
        # Indica que el juego ha finalizado sin encontrar la palabra buscada
        print('Lamentablemente no encontraste la palabra buscada. \nLa palabra era: {}' .format(cad))
```
